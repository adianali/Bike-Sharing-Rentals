```python
##Cleaning Data
#menghapus tabel yang tidak digunakan untuk analisis berdasarkan pertanyaan bisnis
del hr_df
```

```python
# Menghapus beberapa kolom yang tidak akan digunakan
drop_col = ['instant', 'windspeed']

for i in day_df.columns:
    if i in drop_col:
        day_df.drop(labels=i, axis=1, inplace=True)
```

```python
# Mengubah beberapa detail tentang kolom
day_df.rename(columns={
    'dteday': 'dateday',
    'yr': 'year',
    'mnth': 'month',
    'weathersit': 'weather_cond',
    'cnt': 'count'
}, inplace=True)
```

```python
# Mengubah keterangan agar mudah di baca
day_df['month'] = day_df['month'].map({
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
    7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
})
day_df['season'] = day_df['season'].map({
    1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'
})
day_df['weekday'] = day_df['weekday'].map({
    0: 'Sun', 1: 'Mon', 2: 'Tue', 3: 'Wed', 4: 'Thu', 5: 'Fri', 6: 'Sat'
})
day_df['weather_cond'] = day_df['weather_cond'].map({
    1: 'Clear/Partly Cloudy',
    2: 'Misty/Cloudy',
    3: 'Light Snow/Rain',
    4: 'Severe Weather'
})
```

```python
# Mengubah tipe data yang diperlukan
day_df['dateday'] = pd.to_datetime(day_df.dateday)
day_df['season'] = day_df.season.astype('category')
day_df['year'] = day_df.year.astype('category')
day_df['month'] = day_df.month.astype('category')
day_df['holiday'] = day_df.holiday.astype('category')
day_df['weekday'] = day_df.weekday.astype('category')
day_df['workingday'] = day_df.workingday.astype('category')
day_df['weather_cond'] = day_df.weather_cond.astype('category')
day_df.info()
```

```python
day_df.head()
```

```python
## Eksplorasi data day
# Pengelompokan penyewa sepeda (casual dan registered) berdasarkan bulan
day_df.groupby(by='month').agg({
    'count': ['max', 'min', 'mean', 'sum']
})
```

```python
# Pengelompokan penyewa sepeda berdasarkan keadaan cuaca
day_df.groupby(by='weather_cond').agg({
    'count': ['max', 'min', 'mean', 'sum']
})
```

```python
# Pengelompokan penyewa sepeda berdasarkan holiday
day_df.groupby(by='holiday').agg({
    'count': ['max', 'min', 'mean', 'sum']
})
```

```python
# Bandingkan penyewa sepeda pada saat hari kerja vs akhir pekan
day_df.groupby(by='weekday').agg({
    'count':['max','min','mean']
})
```

```python
# Pengelompokan penyewa sepeda berdasarkan workingday
day_df.groupby(by='workingday').agg({
    'count': ['max', 'min', 'mean']
})
```

```python
# Pengelompokan penyewa sepeda berdasarkan season
day_df.groupby(by='season').agg({
    'casual': 'mean',
    'registered': 'mean',
    'count': ['max', 'min', 'mean']
})
```

```python
# Pengelompokan temp, hum, dan humidity berdasarkan season
day_df.groupby(by='season').agg({
    'temp': ['max', 'min', 'mean'],
    'atemp': ['max', 'min', 'mean'],
    'hum': ['max', 'min', 'mean']
})
```

```python
# Hubungan antara penyewa casual, registered, dan count
fig, ax = plt.subplots(figsize=(10,6))
correlation_matrix = day_df.corr(numeric_only=True)
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

sns.heatmap(
    correlation_matrix,
    annot=True,
    mask=mask,
    cmap="viridis",  
    center=0,
    fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

Berdasarkan analisis yang telah dilakukan, dapat disimpulkan sebagai berikut:

- Terdapat korelasi yang sangat tinggi antara atemp dan temp, yakni sebesar 0.99.
- Hubungan antara hum dengan temp dan atemp tergolong lemah, dengan nilai korelasi masing-masing sekitar 0.13 dan 0.14.
- Korelasi antara variabel casual dengan temp dan atemp cukup signifikan, mencapai 0.54, sementara dengan hum bersifat sedikit negatif (-0.08).
- Pola korelasi pada variabel registered mirip dengan casual, dengan korelasi yang moderat terhadap casual, yaitu sekitar 0.40.
- Total penggunaan sepeda (count) memiliki korelasi kuat dengan temp, atemp, casual, dan registered, masing-masing sebesar 0.63, 0.63, 0.67, dan 0.95. Namun, terdapat korelasi negatif yang lemah dengan hum, yakni sekitar -0.10.

```python
# Pertanyaan 1: Apakah ada peran cuaca terhadap jumlah pengguna sepeda?
plt.figure(figsize=(10, 6))
sns.barplot(
    x='weather_cond',
    y='count',
    data=day_df,
    palette='Paired' 
)
plt.title('Pengguna Sepeda berdasarkan Kondisi Cuaca')
plt.xlabel('Kondisi Cuaca')
plt.ylabel('Jumlah Pengguna Sepeda')
plt.show()
```

```python
# Pertanyaan 2: Bagaimana kondisi yang tampak ketika sepeda digunakan pada workingday, holiday, dan weekday?

fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(15, 10))

# Subplot 1
sns.barplot(
    x='workingday',
    y='count',
    data=day_df,
    ax=axes[0],
    palette='bright'
)
axes[0].set_title('Pengguna Sepeda berdasarkan Hari Kerja')
axes[0].set_xlabel('Hari Kerja')
axes[0].set_ylabel('Jumlah Pengguna Sepeda')

# Subplot 2
sns.barplot(
    x='holiday',
    y='count',
    data=day_df,
    ax=axes[1],
    palette='bright'
)
axes[1].set_title('Pengguna Sepeda berdasarkan Hari Libur')
axes[1].set_xlabel('Hari Libur')
axes[1].set_ylabel('Jumlah Pengguna Sepeda')

# Subplot 3
sns.barplot(
    x='weekday',
    y='count',
    data=day_df,
    ax=axes[2],
    palette='bright'
)
axes[2].set_title('Pengguna Sepeda berdasarkan Hari dalam Seminggu')
axes[2].set_xlabel('Hari dalam Seminggu')
axes[2].set_ylabel('Jumlah Pengguna Sepeda')

plt.tight_layout()
plt.show()
```

```python
# Pertanyaan 3: Bagaimana tren penggunaan sepeda dalam setahun, apakah tahun 2011 atau 2012 yang lebih unggul?
day_df['month'] = pd.Categorical(day_df['month'], categories=
    ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'],
    ordered=True)

monthly_counts = day_df.groupby(by=["month","year"]).agg({
    "count": "sum"
}).reset_index()

sns.lineplot(
    data=monthly_counts,
    x="month",
    y="count",
    hue="year",
    palette="rocket",
    marker="o")

plt.title("Total sepeda yang disewakan berdasarkan Bulan dan tahun")
plt.xlabel(None)
plt.ylabel(None)
plt.legend(title="Tahun", loc="upper right")
plt.tight_layout()
plt.show()
```

```python
# Pertanyaan 4: Bagaimana variabel seperti temp dan atemp serta humidity dapat mempengaruhi jumlah total pengguna sepeda (baik Casual ataupun Registered)?
plt.figure(figsize=(14, 6))
plt.subplot(1, 3, 1)
sns.scatterplot(
    x='temp',
    y='count',
    data=day_df,
    alpha=0.5
)
plt.title('Temperature vs Count')

plt.subplot(1, 3, 2)
sns.scatterplot(
    x='atemp',
    y='count',
    data=day_df,
    alpha=0.5
)
plt.title('Feels Like Temperature vs Count')

plt.subplot(1, 3, 3)
sns.scatterplot(
    x='hum',
    y='count',
    data=day_df,
    alpha=0.5
)
plt.title('Humidity vs Count')
plt.show()
```

```python
# Pertanyaan 5: Apakah ada pengaruh musim terhadap penyewa sepeda berdasarkan pelanggan Casual dan Registered?
seasonal_usage = day_df.groupby('season')[['registered', 'casual']].sum().reset_index()

plt.figure(figsize=(10, 6))
plt.bar(
    seasonal_usage['season'],
    seasonal_usage['registered'],
    label='Registered',
    color='tab:blue'
)

plt.bar(
    seasonal_usage['season'],
    seasonal_usage['casual'],
    label='Casual',
    color='tab:red'
)

plt.xlabel(None)
plt.ylabel(None)
plt.title('Penyewaan sepeda berdasarkan musim')
plt.legend()
plt.show()
```

# Kesimpulan

# Pertanyaan 1:
Dapat dengan jelas dilihat bahwa terdapat hubungan antara kondisi cuaca dan jumlah penyewa sepeda. Para penyewa sepeda cenderung menyukai cuaca yang cerah atau sedikit mendung, dengan nilai kuartil ketiga (Q3) di atas 6000, kuartil pertama (Q1) di bawah 4000, dan median hampir mencapai 5000. Kedua, cuaca berkabut atau mendung menduduki posisi kedua, dengan Q3 di atas 5000, Q1 di bawah 3000, dan median di atas 4000. Terakhir, cuaca sedikit bersalju atau hujan kurang diminati, dengan Q3 di atas 2000, Q1 di bawah 1000, dan median di atas 1000. Data cuaca saat badai besar tidak tersedia karena tidak memungkinkan untuk bersepeda dalam kondisi cuaca ekstrem tersebut.

# Pertanyaan 2:
Kondisi terlihat terbagi menjadi tiga bagian. Jumlah pengguna sepeda lebih tinggi pada hari kerja (workingday) dibandingkan dengan hari libur (non-workingday), walaupun perbedaannya tidak terlalu signifikan, tetapi masih jelas di atas 4000 pengguna. Pada hari libur, non-holiday unggul dengan perbedaan yang cukup besar dibandingkan dengan hari libur (holiday). Ini mungkin menunjukkan bahwa pengguna lebih suka menyewa sepeda saat hari biasa. Berdasarkan hari dalam seminggu (weekday), Jumat (Fri) menduduki peringkat pertama dengan mean 4690.288462, sementara Minggu (Sun) berada di peringkat terakhir dengan mean 4228.828571.

# Pertanyaan 3:
Visualisasi menunjukkan perbedaan puncak dan dasar bulan antara tahun 2011 dan 2012. Puncak penyewaan pada tahun 2011 terjadi pada bulan Juni, sedangkan dasarnya terjadi pada bulan Januari. Pada tahun 2012, puncaknya terjadi pada bulan September dengan dasar yang sama pada bulan Januari. Jumlah total sepeda yang disewakan pada tahun 2012 secara keseluruhan lebih tinggi daripada pada tahun 2011.

# Pertanyaan 4:
Dari tiga visualisasi yang ditampilkan, dapat disimpulkan bahwa suhu (temp dan atemp) memiliki korelasi positif dengan jumlah penyewa sepeda (count), menunjukkan bahwa kenaikan suhu berhubungan dengan peningkatan penggunaan sepeda. Sebaliknya, kelembaban (humidity) memiliki korelasi negatif yang sangat minimal tetapi masih dapat mempengaruhi penurunan penggunaan sepeda ketika kelembaban meningkat.

# Pertanyaan 5:
Hasil visualisasi terakhir cukup mengejutkan, di mana musim dingin (Winter) tidak menjadi yang terakhir, melainkan musim semi (Spring). Faktor-faktor seperti kondisi cuaca, suhu udara, hari kerja, hari libur, dan lain-lain dapat mempengaruhi hasil ini.
